# **simple open-source image captioning models**

## 1. BLIP (Bootstrapping Language-Image Pretraining)

* Hugging Face: Salesforce/blip-image-captioning-base

* Works well on Colab (small enough for T4).

* Very simple API in 🤗 Transformers.

* mainly trained for image captioning (describing what’s in an image)

## 2. BLIP-2 (VQA capable)

* Hugging Face: Salesforce/blip2-flan-t5-xl

* Supports Visual Question Answering (VQA) in addition to captioning.

* Works in Colab, but a bit heavier than BLIP.

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import requests
```

```python
# Load model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
# Load image
url = "https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg"
image = Image.open(requests.get(url, stream=True).raw)
```

```python
path = "/content/label_0_1000.jpeg"
image2 = Image.open(path).convert("RGB")
```

```python
# Generate caption
inputs = processor(image2, return_tensors="pt")
out = model.generate(**inputs)
print(processor.decode(out[0], skip_special_tokens=True))
```

```python
from transformers import Blip2Processor, Blip2ForConditionalGeneration
import torch
```
